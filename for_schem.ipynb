```python
import numpy as np
```

```python
eps_0 = 8.85418781762039e-12  # Vacuum permittivity in F/m
mu_0 = 4e-7 * np.pi # Vacuum permeability in H/m
print(mu_0)
c = 1 / np.sqrt(eps_0 * mu_0)  # Speed of light in m/s]
eta_0 = np.sqrt(mu_0 / eps_0)  # Intrinsic impedance of free space in ohms
print(f"Speed of light: {c} m/s")
print(f"Intrinsic impedance: {eta_0} ohms")
```

$\alpha = \dfrac{\eta}{X_s}$, где $X_s$ - мнимая часть импеданса импедансного листа

```python
f = 5e9  # Frequency in Hz
w = 2 * np.pi * f  # Angular frequency in rad/s
k = w / c  # Wave number in rad/m
def calc_nominal_from_alpha(alpha, f=1e9):
    w = 2 * np.pi * f
    k = w / c
    x_s = eta_0/alpha
    print(x_s, "это x_s")
    if alpha > 0:
        print(f"inductive sheet {x_s/w} H")
    if alpha < 0:
        print(f"capacitive sheet {-1/(x_s*w)} F")
```

```python
alpha = 1.5
l_best = (np.atan((alpha - np.sqrt(alpha**2 + 4))/2)+np.pi)/k
coef = np.sqrt(1/((alpha*np.sin(k*l_best)+np.cos(k*l_best))**2+(np.sin(k*l_best))**2))
lamb = c/f
print(f"Frequency: {f/1e9} GHz, Wavelength: {lamb} m")
print('wavelength/4', lamb/4)
print(f"Optimum length for alpha={alpha}: {l_best} m, {l_best/lamb*360} degrees") 
print(f"amp coef: {coef}")
```

```python
calc_nominal_from_alpha(alpha, f)
```
